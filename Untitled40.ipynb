```python
# ===== Download dataset from Kaggle
!kaggle datasets download somnambwl/bookcrossing-dataset -q
!unzip -o bookcrossing-dataset.zip -d bookcrossing

# Load data into pandas
books = pd.read_csv('bookcrossing/Books.csv', sep=';', encoding='latin-1', on_bad_lines='skip', low_memory=False)
ratings = pd.read_csv('bookcrossing/Ratings.csv', sep=';', encoding='latin-1', on_bad_lines='skip', low_memory=False)
users = pd.read_csv('bookcrossing/Users.csv', sep=';', encoding='latin-1', on_bad_lines='skip', low_memory=False)  # optional

print("Books:", books.shape, "Ratings:", ratings.shape)
```

```python
def _lower_nospace(s): return s.strip().lower().replace('-', '').replace('_',' ').replace('  ',' ')

def standardize_book_cols(df):
    colmap = {}
    for c in df.columns:
        k = _lower_nospace(c)
        if k in ['isbn']: colmap[c] = 'isbn'
        elif k in ['book title','title']: colmap[c] = 'title'
        elif k in ['book author','author']: colmap[c] = 'author'
        elif k in ['publisher']: colmap[c] = 'publisher'
        elif k in ['yearofpublication','year','publicationyear']: colmap[c] = 'year'
    return df.rename(columns=colmap)[['isbn','title']].dropna()

def standardize_rating_cols(df):
    colmap = {}
    for c in df.columns:
        k = _lower_nospace(c)
        if k in ['userid','user id','user']: colmap[c] = 'user_id'
        elif k in ['isbn']: colmap[c] = 'isbn'
        elif k in ['bookrating','rating','book rating']: colmap[c] = 'rating'
    return df.rename(columns=colmap)[['user_id','isbn','rating']].dropna()

books_std = standardize_book_cols(books)
ratings_std = standardize_rating_cols(ratings)

print("✅ Columns standardized")
```

```python
# Keep only explicit ratings (>0)
ratings_explicit = ratings_std[ratings_std['rating'] > 0]

# Users with >=50 ratings
user_counts = ratings_explicit.groupby('user_id')['rating'].count()
users_kept = user_counts[user_counts >= 50].index
df = ratings_explicit[ratings_explicit['user_id'].isin(users_kept)]

# Merge with book titles
df = df.merge(books_std[['isbn','title']], on='isbn', how='inner')

# Books with >=20 ratings
book_counts = df.groupby('title')['rating'].count()
books_kept = book_counts[book_counts >= 20].index
df = df[df['title'].isin(books_kept)]

print(f"Filtered dataset: {df['user_id'].nunique()} users, {df['title'].nunique()} books, {len(df)} ratings")
```

```python
book_user_mat = df.pivot_table(index='title', columns='user_id', values='rating', aggfunc='mean').fillna(0)
sparse_matrix = csr_matrix(book_user_mat.values)

model = NearestNeighbors(metric='cosine', algorithm='brute')
model.fit(sparse_matrix)

titles = book_user_mat.index.tolist()
title_to_idx = {t:i for i,t in enumerate(titles)}

print("✅ KNN model trained")
```

```python
def get_recommends(book_title, n=5):
    if book_title not in title_to_idx:
        candidates = [t for t in titles if book_title.lower() in t.lower()]
        suggestion = " Did you mean? " + "; ".join(candidates[:10]) if candidates else ""
        raise ValueError(f"Title not found: {book_title}.{suggestion}")
    idx = title_to_idx[book_title]
    distances, indices = model.kneighbors(sparse_matrix[idx], n_neighbors=n+1)
    recs = [[titles[i], float(d)] for d,i in zip(distances[0][1:], indices[0][1:])]
    return [book_title, recs]

def search_titles(q, k=10):
    return [t for t in titles if q.lower() in t.lower()][:k]

print("✅ get_recommends() and search_titles() ready")
```

```python
query = "The Queen of the Damned (Vampire Chronicles (Paperback))"
try:
    res = get_recommends(query)
    print("Query:", res[0])
    for rec, dist in res[1]:
        print(f"  – {rec} (distance={dist})")
except ValueError as e:
    print(e)
    print("Try search_titles('vampire') for suggestions")
```
